```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_excel("PYL_2020_2023_LIMPIO - Finanzas.xlsx")
df
```

```python
df.rename({
    'Net proceeds of sales (for NNS) (3030)':'Ventas_Brutas',
    'Net net proceeds of sales (3060)': 'Ventas_Netas',
    'Total variable expenses (for NNS) (Margin)(3150)':'Costo_de_Ventas', 
    'Total fixed product expenses (for NNS) (3270)': 'Gastos Fijos_de_Fabrica',
    'Total product fixed marketing expenses (3225)':'Gastos_de_Marketing',
    'Total fixed overhead expenses (3390)':'Gastos_Generales_Fijos',
    'Total allowances':'Descuentos_distribuidores_desempeño',
    'Consolidated Operating Profit - COP1 ': 'Resultado_Operacional',
    'Fiscal Year': 'Anio',
    'Posting period': 'Mes',
    'Group': 'Unidad'
}, axis=1, inplace=True)
```

```python
df
```

```python
# Seleccionar los datos relevantes
datos = df[['Anio', 'Ventas_Netas', 'Gastos_de_Marketing']]

# Calcular la correlación por año
correlacion_por_anio = datos.groupby('Anio').apply(lambda x: x.corr().loc['Ventas_Netas', 'Gastos_de_Marketing'])

print("Correlación entre gastos de marketing y ventas por año:")
print(correlacion_por_anio)

# Graficar la correlación por año
plt.figure(figsize=(10, 6))
sns.barplot(x=correlacion_por_anio.index, y=correlacion_por_anio.values)
plt.xlabel('Anio')
plt.ylabel('Correlación')
plt.title('Correlación entre Gastos de Marketing y Ventas por Año')
plt.show()

# Visualizar la correlación con un diagrama de dispersión por año
for year in datos['Anio'].unique():
    subset = datos[datos['Anio'] == year]
    sns.scatterplot(data=subset, x='Ventas_Netas', y='Gastos_de_Marketing', label=year)

plt.title('Correlación entre Gastos de Marketing y Ventas por Año')
plt.xlabel('Ventas')
plt.ylabel('Gastos de Marketing')
plt.legend(title='Año')
plt.show()

# Visualizar la matriz de correlación usando un heatmap por año
for year in datos['Anio'].unique():
    subset = datos[datos['Anio'] == year]
    correlacion = subset.corr()
    plt.figure(figsize=(6, 4))
    sns.heatmap(correlacion, annot=True, cmap='coolwarm')
    plt.title(f'Matriz de Correlación - Año {year}')
    plt.show()
```

```python
# Seleccionar los datos relevantes
datos = df[['Gastos_de_Marketing','Descuentos_distribuidores_desempeño']]

# Calcular la correlación
correlacion = datos.corr()
print("Correlación entre Descuentos por desempeño y Gastos_de_Marketing:")
print(correlacion)

# Visualizar la correlación
sns.scatterplot(data=datos, x='Gastos_de_Marketing', y='Descuentos_distribuidores_desempeño')
plt.title('Correlación entre Descuentos por desempeño y Gastos_de_Marketing')
plt.xlabel('Gastos_de_Marketing')
plt.ylabel('Descuentos por desempeño')
plt.show()

# Visualizar la matriz de correlación usando un heatmap
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Seleccionar las columnas de interés
columns_of_interest = ['Anio', 'Ventas_Netas', 'Descuentos_distribuidores_desempeño']
data_subset = df[columns_of_interest]

# Asegurarse de que no hay valores nulos en las columnas de interés
data_subset = data_subset.dropna()

# Calcular la correlación por año
correlation_by_year = data_subset.groupby('Anio').apply(
    lambda x: x.corr().loc['Ventas_Netas', 'Descuentos_distribuidores_desempeño']
)

# Mostrar la correlación por año
print(correlation_by_year)

# Graficar la correlación por año
plt.figure(figsize=(10, 6))
sns.barplot(x=correlation_by_year.index, y=correlation_by_year.values)
plt.xlabel('Ventas Netas')
plt.ylabel('Correlacion')
plt.title('Correlación entre Descuentos por desempeño y ventas')
plt.show()
```

```python
# Configurar pandas para mostrar los números completos en lugar de notación científica
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# Seleccionar las columnas específicas
columnas_especificas = ['Ventas_Brutas', 
                        'Costo_de_Ventas', 
                        'Gastos_Generales_Fijos']

# Dividir los valores por 1,000 para que salgan en miles
df_miles = df[columnas_especificas] / 1000

# Aplicar el método describe() a las columnas seleccionadas
descripcion_especifica = df[columnas_especificas].describe()

# Mostrar la descripción
print(descripcion_especifica)
```

```python
# Agrupar los datos por Unidad de Negocio y Año, y sumar las Ventas Netas
df_grouped = df.groupby(['Unidad', 'Anio', 'Mes'])['Ventas_Netas'].sum().reset_index()

# Crear el gráfico de tendencia
plt.figure(figsize=(12, 6))
for unidad in df_grouped['Unidad'].unique():
    unidad_data = df_grouped[df_grouped['Unidad'] == unidad]
    plt.plot(unidad_data['Anio'].astype(str) + '-' + unidad_data['Mes'].astype(str), unidad_data['Ventas_Netas'], label=unidad)

plt.title('Tendencia de Ventas Netas por Unidad de Negocio')
plt.xlabel('Periodo')
plt.ylabel('Ventas Netas')
plt.legend(title='Unidad de Negocio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular Margen Bruto y Margen Neto
df['Margen_Bruto'] = df['Ventas_Netas'] - df['Costo_de_Ventas']
df['Margen_Neto'] = df['Resultado_Operacional'] / df['Ventas_Netas']

# Resumen de rentabilidad por Unidad de Negocio
rentabilidad = df.groupby('Unidad')[['Ventas_Netas', 'Resultado_Operacional', 'Margen_Neto']].mean()
print(rentabilidad)
```

```python
# Calcular Ratios
df['Ratio_Gastos_Marketing'] = df['Gastos_de_Marketing'] / df['Ventas_Netas']
df['Ratio_Gastos_Generales'] = df['Gastos_Generales_Fijos'] / df['Ventas_Netas']

# Resumen de ratios por Unidad de Negocio
ratios = df.groupby('Unidad')[['Ratio_Gastos_Marketing', 'Ratio_Gastos_Generales']].mean()
print(ratios)
```

```python
# Distribución de Ventas Netas
plt.figure(figsize=(12, 6))
sns.histplot(df['Ventas_Netas'], kde=True)
plt.title('Distribución de Ventas Netas')
plt.xlabel('Ventas Netas')
plt.ylabel('Frecuencia')
plt.show()
```

Forma de la Distribución:

La distribución parece ser asimétrica hacia la derecha (sesgo positivo). Esto significa que la mayoría de las ventas netas están concentradas en los rangos más bajos, pero hay algunos casos con ventas netas significativamente más altas.
La mayor parte de los datos se encuentra en el rango de ventas netas de 0 a 2 millones, con un número decreciente de instancias a medida que aumentan las ventas netas

Hay un pico notable en el primer bin (0 a 2 millones), lo que indica que muchas observaciones tienen ventas netas bajas.
Hay otros picos menores alrededor de los 6 millones, lo que podría sugerir que hay algunas unidades de negocio o períodos específicos con ventas netas más altas en comparación con el resto.

La forma de la curva respalda la observación de que los datos están sesgados hacia la derecha.
